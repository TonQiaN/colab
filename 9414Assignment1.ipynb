```python
# selected feature
''' Index(['CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)',
       'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)',
       'T', 'RH', 'AH'],
      dtype='object') '''
# print(variables)
variables = ['PT08.S1(CO)', 'C6H6(GT)', 'PT08.S2(NMHC)',
       'NOx(GT)', 'NO2(GT)', 'PT08.S4(NO2)'
       ]
variables_drop = [columns for columns in df.columns if columns not in variables]
y = df['CO(exceeds)']
x = df.drop(columns=variables_drop)
print(x)
print(y)
```

```python
from sklearn.preprocessing import StandardScaler

# åˆå§‹åŒ– StandardScaler
scaler = StandardScaler()

# å¯¹ x è¿›è¡Œæ ‡å‡†åŒ–å¤„ç†
normalisation = scaler.fit_transform(x)

# å°†æ ‡å‡†åŒ–åŽçš„æ•°æ®è½¬æ¢å›ž DataFrame
normalisation_table = pd.DataFrame(normalisation, columns=x.columns)
x = normalisation

print(x)
print(y)
print(normalisation_table)

# from sklearn.preprocessing import MinMaxScaler
# scaler = MinMaxScaler()
# normalisation = scaler.fit_transform(x)
# normalisation_table = pd.DataFrame(normalisation, columns=x.columns)
# x = normalisation
# # y = y.to_numpy()
# print(x)
# print(y)
# print(normalisation_table)
```

```python
from sklearn.model_selection import train_test_split
# ç¬¬ä¸€æ­¥ï¼šå°†æ•°æ®é›†æŒ‰ 6:4 åˆ†ä¸ºè®­ç»ƒé›†å’Œä¸´æ—¶é›†
training_data, temp_data, y_train, y_temp = train_test_split(x, y, test_size=0.4, random_state=42)

# ç¬¬äºŒæ­¥ï¼šå°†ä¸´æ—¶é›†æŒ‰ 2:1 åˆ†ä¸ºéªŒè¯é›†å’Œæµ‹è¯•é›†
validation_data, test_data, y_validation, y_test = train_test_split(temp_data, y_temp, test_size=0.5, random_state=42)

# æ‰“å°æ•°æ®é›†çš„å¤§å°
print(f'Total samples: {x.shape[0]}')
print(f'Training samples: {training_data.shape[0]}')
print(f'Validation samples: {validation_data.shape[0]}')
print(f'Test samples: {test_data.shape[0]}')

print(f'Total samples: {y.shape[0]}')
print(f'Training labels: {y_train.shape[0]}')
print(f'Validation labels: {y_validation.shape[0]}')
print(f'Test labels: {y_test.shape[0]}')

# total_samples = x.shape[0]
# training_size = int(total_samples * 0.7)
# validation_size = int(total_samples * 0.2)
# # test_size = total_samples - training_size - validation_size

# training_data = x[:training_size]
# validation_data = x[training_size:training_size + validation_size]
# test_data = x[training_size + validation_size:]

# y_train = y[:training_size]
# y_validation = y[training_size:training_size + validation_size]
# y_test = y[training_size + validation_size:]

# print(total_samples,training_data.shape[0], test_data.shape[0], validation_data.shape[0])
# print(total_samples,y_train.shape[0], y_test.shape[0], y_validation.shape[0])
# print(f'{training_data}\n\n, {test_data}\n\n, {validation_data}\n\n')
# print(f'{y_train}\n\n, {y_test}\n\n, {y_validation}')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

output_size_to_next_layer = 24
regularization_number = 0.1
c = 16

model = Sequential([
           Dense(output_size_to_next_layer, activation='relu', input_shape=(training_data.shape[1],)),
           Dropout(regularization_number),
           Dense(c, activation='relu'),
           Dropout(regularization_number),
           Dense(1, activation='sigmoid')
           ])
alpha = 0.001
model.compile(optimizer=Adam(learning_rate=alpha), loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
print(training_data.shape[0] / 10)
```

```python
history = model.fit(training_data, y_train, epochs=100, batch_size=32, validation_data=(validation_data, y_validation))
history_df = pd.DataFrame(history.history)

# loss, accuracy = model.evaluate(test_data, y_test)
# print(f"Test Loss: {loss}")
# print(f"Test Accuracy: {accuracy}")

# # è¿›è¡Œé¢„æµ‹
# y_pred = model.predict(test_data)
# print(y_pred)
```

```python
# ç»˜åˆ¶ loss æ›²çº¿
plt.figure(figsize=(12, 6))
plt.plot(history_df['loss'], label='Training Loss')
plt.plot(history_df['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ç»˜åˆ¶ accuracy æ›²çº¿
plt.figure(figsize=(12, 6))
plt.plot(history_df['accuracy'], label='Training Accuracy')
plt.plot(history_df['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score

# ç”¨æ¨¡åž‹é¢„æµ‹æµ‹è¯•é›†çš„æ ‡ç­¾
y_test_pred_prob = model.predict(test_data)
y_test_pred = (y_test_pred_prob > 0.5).astype(int)

# è®¡ç®—æ··æ·†çŸ©é˜µ
cm = confusion_matrix(y_test, y_test_pred)

# æ‰“å°æ··æ·†çŸ©é˜µ
print("Confusion Matrix:\n", cm)

# ç»˜åˆ¶æ··æ·†çŸ©é˜µ
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# è®¡ç®— Accuracy å’Œ Precision
accuracy = accuracy_score(y_test, y_test_pred)
precision = precision_score(y_test, y_test_pred)

# æ‰“å° Accuracy å’Œ Precision
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")

# ä½ çš„å‡†ç¡®çŽ‡å’Œç²¾ç¡®åº¦æ•°æ®
accuracy_precision_data = {
    'Metric': ['Accuracy', 'Precision', 'Number of Samples'],
    'Value': [f'{accuracy*100:.2f}%', f'{precision*100:.2f}%', y_test.shape[0]]
}

# åˆ›å»ºå›¾å½¢å’Œè½´
fig, axs = plt.subplots(figsize=(10, 0.4))

# ç»˜åˆ¶å‡†ç¡®çŽ‡å’Œç²¾ç¡®åº¦è¡¨
accuracy_precision_df = pd.DataFrame(accuracy_precision_data)
axs.axis('off')
axs.table(cellText=accuracy_precision_df.values,
             cellLoc='center')

# è°ƒæ•´å¸ƒå±€
plt.show()
```

```python
from sklearn.model_selection import train_test_split
# ç¬¬ä¸€æ­¥ï¼šå°†æ•°æ®é›†æŒ‰ 7:3 åˆ†ä¸ºè®­ç»ƒé›†å’Œä¸´æ—¶é›†
training_data, temp_data, y_train, y_temp = train_test_split(x, y, test_size=0.4, random_state=42)

# ç¬¬äºŒæ­¥ï¼šå°†ä¸´æ—¶é›†æŒ‰ 2:1 åˆ†ä¸ºéªŒè¯é›†å’Œæµ‹è¯•é›†
validation_data, test_data, y_validation, y_test = train_test_split(temp_data, y_temp, test_size=0.5, random_state=42)

# æ‰“å°æ•°æ®é›†çš„å¤§å°
print(f'Total samples: {x.shape[0]}')
print(f'Training samples: {training_data.shape[0]}')
print(f'Validation samples: {validation_data.shape[0]}')
print(f'Test samples: {test_data.shape[0]}')

print(f'Total samples: {y.shape[0]}')
print(f'Training labels: {y_train.shape[0]}')
print(f'Validation labels: {y_validation.shape[0]}')
print(f'Test labels: {y_test.shape[0]}')

# total_samples = x.shape[0]
# training_size = int(total_samples * 0.7)
# validation_size = int(total_samples * 0.2)
# # test_size = total_samples - training_size - validation_size

# training_data = x[:training_size]
# validation_data = x[training_size:training_size + validation_size]
# test_data = x[training_size + validation_size:]

# y_train = y[:training_size]
# y_validation = y[training_size:training_size + validation_size]
# y_test = y[training_size + validation_size:]

# print(total_samples,training_data.shape[0], test_data.shape[0], validation_data.shape[0])
# print(total_samples,y_train.shape[0], y_test.shape[0], y_validation.shape[0])
# print(f'{training_data}\n\n, {test_data}\n\n, {validation_data}\n\n')
# print(f'{y_train}\n\n, {y_test}\n\n, {y_validation}')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

output_size_to_next_layer = 24
regularization_number = 0.4
c = 16

model = Sequential([
           Dense(output_size_to_next_layer, activation='relu', input_shape=(training_data.shape[1],)),
           Dropout(regularization_number),
           Dense(c, activation='relu'),
           Dropout(regularization_number),
           Dense(1, activation='linear')
           ])
alpha = 0.001
model.compile(optimizer=Adam(learning_rate=alpha), loss='mean_squared_error', metrics=['mean_absolute_error'])
model.summary()
print(training_data.shape[0] / 10)
```

```python
history = model.fit(training_data, y_train, epochs=100, batch_size=32, validation_data=(validation_data, y_validation))
history_df = pd.DataFrame(history.history)
```

```python
# ç»˜åˆ¶ Loss æ›²çº¿
plt.figure(figsize=(12, 6))
plt.plot(history_df['loss'], label='Training Loss')
plt.plot(history_df['val_loss'], label='Validation Loss')
plt.title('Loss Curve for the Regression Task')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ç”¨æ¨¡åž‹é¢„æµ‹éªŒè¯é›†çš„ NOx(GT) å€¼
y_val_pred = model.predict(validation_data)

# åˆ›å»ºæ–°çš„ x è½´ç´¢å¼•
x_index = np.arange(len(y_validation))

# ç»˜åˆ¶ Estimated and Actual NOx(GT) for the Validation Set
plt.figure(figsize=(12, 6))
plt.plot(x_index, y_validation, label='Actual NOx(GT)')
plt.plot(x_index, y_val_pred, label='Estimated NOx(GT)')
plt.title('Estimated and Actual NOx(GT) for the Validation Set')
plt.xlabel('Sample Index')
plt.ylabel('NOx(GT)')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
# è®¡ç®— RMSE å’Œ MAE
rmse = np.sqrt(mean_squared_error(y_validation, y_val_pred))
mae = mean_absolute_error(y_validation, y_val_pred)

# æ‰“å° RMSE å’Œ MAE
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f'Number of Samples: {y_test.shape[0]}')

# ä½ çš„å‡†ç¡®çŽ‡å’Œç²¾ç¡®åº¦æ•°æ®
accuracy_precision_data = {
    'Metric': ['RMSE', 'MAE', 'Number of Samples'],
    'Value': [f'{rmse:.2f}', f'{mae:.2f}', y_test.shape[0]]
}

# åˆ›å»ºå›¾å½¢å’Œè½´
fig, axs = plt.subplots(figsize=(10, 0.4))

# ç»˜åˆ¶å‡†ç¡®çŽ‡å’Œç²¾ç¡®åº¦è¡¨
accuracy_precision_df = pd.DataFrame(accuracy_precision_data)
axs.axis('off')
axs.table(cellText=accuracy_precision_df.values,
             cellLoc='center')

# è°ƒæ•´å¸ƒå±€
plt.show()
```
